## Funciones trascendentes

Las funciones trascendentes no pueden ser expresadas con polinomios. Solo con funciones tales como las trigonométricas, exponenciales, logaritmos

## Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo, la función escalón de Heaviside:

$
f(x)= \left\{ \begin{array}{lcc}
             0 &   para  & x < 0 \\            
             \\ 1 &  para  & x \geq 0
             \end{array}
   \right.
$


```python
def H(x):
  Y = np.zeros(len(x))
  for idx, x in enumerate(x):
    if x >= 0:
      Y[idx] = 1
  return Y

N = 1000
x = np.linspace(-10,10,num=N)
y = H(x)

plt.plot(x,y);
```

```python
def f(x):
  return np.absolute(x)

x = np.linspace(-10,10, num=1000)

y=f(x)

plt.plot(x, y);
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x + c)
yx = f(x - c)
fig, ax = plt.subplots(1,3, figsize=(10,10))
ax[0].plot(x,y)
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')

ax[1].plot(x,yx)
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r');

ax[2].plot(x,f(x))
ax[2].grid()
ax[2].axhline(y=0, color='r')
ax[2].axvline(x=0, color='r');
```

Siendo $c$ una constante mayor que cero, entonces la gráfica:

- $y=cf(x)$ alarga la gráfica verticalmente en un factor $c$.
- $y=\frac{1}{c}f(x)$ comprime la gráfica verticalmente en un factor $c$.
- $y=f(x*c)$ comprime la gráica horizontalmente en un factor de $c$.
- $y=\frac{1}{c}f(x)$ alarca la gráica horizontalmente en un factor de $c$.

```python
N = 1000
def f(x):
  return np.sin(x)

c = 2
x = np.linspace(-15,15, num= N)
y = c*f(x)
fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r');
```

## Reflexiones

+ $y = -f(x)$ refleja la gráfica respecto al eje x.
+ $y = f(-x)$ refleja la gráfica respecto al eje y.

```python
N = 1000
def f(x):
  return x**3

x = np.linspace(-10,10, num= N)
y = f(x)
functions = [y, -f(x), f(-x)] 
titles = ['f(x)', '-f(x)', 'f(-x)'] 
fig, ax = plt.subplots(1,3, figsize=(10,10))
for i in range(3):
  y = functions[i]
  ax[i].plot(x,y)
  ax[i].set_title(titles[i])
  ax[i].grid()
  ax[i].axhline(y=0, color='r')
  ax[i].axvline(x=0, color='r');
```

## Funciones pares e impares

+ Función par -> $f(-x) = f(x)$
+ Función impar -> $f(-x) = -f(x)$

```python
def f(x, e):
  return x**e

x = 3
print("Es par -> {0}".format(f(x,3) == f(-x,3)))
print("Es impar -> {0}".format(-f(x,3) == f(-x,3)))
print("*"*15)
print("Es par -> {0}".format(f(x,2) == f(-x,2)))
print("Es impar -> {0}".format(-f(x,2) == f(-x,2)))
```

```python
x = np.linspace(-20,20, num = 1000)
fig, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].plot(x,f(x,3))
ax[0].grid();
ax[0].set_title("Impar f(x) = x**3")
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r');
ax[1].plot(x,f(x,2))
ax[1].grid();
ax[1].set_title("Par f(x) = x**2");
```

## Funciones concavas y convexas

+ $f^{\prime\prime}(x)$ > 0 entonces $f(x)$ es convexa
+ $f^{\prime\prime}(x)$ < 0 entonces $f(x)$ es cóncava

```python
def f(x):
  return 2*(x-2)**2 + 1
def g(x):
  return -2*(x-2)**2 + 1

x = np.linspace(-20,20, num = 1000)
fig, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].plot(x,f(x))
ax[0].grid();
ax[0].set_title("Convexa")
ax[1].plot(x,g(x))
ax[1].grid();
ax[1].set_title("Cóncava");
```

## Funciones de activación

```python
N = 1000
x = np.linspace(-5,5, num=N)
```

### Función lineal
$$y=mx+b$$

```python
def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

### Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N = 1000

y = H(x)
plt.plot(x,y)
plt.grid()
```

### Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    
N = 1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

### Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
  return np.tanh(x)
    
N = 1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

### Función ReLU

$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)
    
N = 1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

**¿Para qué se utilizan las funciones?**

+ $\textbf{Función lineal}$: mantener valores a lo largo de un proceso, predecir venta

+ $\textbf{Función escalón o de Heaviside}$: clasificaciones categóricas, ejemplo binarias

+ $\textbf{Función sigmoide}$: regresión logística, para probabilidades

+ $\textbf{Función tangente hiperbólica}$: para escalamiento

+ $\textbf{Función ReLU}$: simulación de “neuronas muertas”
