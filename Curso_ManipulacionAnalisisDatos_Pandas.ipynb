## Lidiar con datos duplicados en Pandas

```python
df = pd.DataFrame({'a': ['w'] * 4 + ['x'] * 3 + ['y'] * 2 + ['z']+['v'], 
                   'b': [1, 1, 1, 1, 2, 2, 2, 3, 3, 4,5]})

df
```

```python
print("Duplicados")
print(df.duplicated()) # Ver duplicados
print("Duplicados - Marca solo la primera ocurrencia")
print(df.duplicated(keep='first')) # marca solo la primera ocurrencia  
print("Duplicados - Marca solo la última ocurrencia")
print(df.duplicated(keep='last')) # marca solo la última ocurrencia
print("Duplicados - Filtra aquellos casos que no tienen duplicados")
print(df[~ df.duplicated()]) # Filtra aquellos casos que no tienen duplicados, el ~ es negación
print("Duplicados - Marca todos los registros duplicados")
print(df.duplicated(keep=False)) # Marca todos los registros duplicados
print("Duplicados - Descarta los registros duplicados")
print(df.drop_duplicates()) # descarta los registros duplicados, tiene por defecto el keep = first
print("Duplicados - Elimina los duplicados teniendo en cuenta una sola columna")
print(df.drop_duplicates(['a'],keep='last')) # elimina los duplicados teniendo en cuenta una sola columna
```

```python
# Convierte dt a variable tipo tiempo
df_temp['year'] = pd.to_datetime(df_temp.dt)
df_temp['year']
```

```python
# Un filtro
idx = df_temp.year > pd.to_datetime('1970-01-01')
df_temp = df_temp[idx]
df_temp
```

```python
df_temp_avg = df_temp.groupby(['Country', pd.Grouper(key='year', freq='1Y')]).agg({'AverageTemperature':[np.mean,np.median]})
df_temp_avg
```

```python
df_temp_avg.xs('Mexico')['AverageTemperature']
```

```python
df_temp_avg.xs('Mexico')['AverageTemperature'].plot();
```

```python
df_tmed = df_temp_avg['AverageTemperature'][['median']].reset_index()
df_tmed
```

```python
df_tmed['date'] = df_tmed['year'].dt.year
df_tmed['date'] 
```

```python
df_tmed.rename(columns={'median':'temperature'}, inplace=True)
df_tmed
```

```python
df_t_pivot = df_tmed.pivot_table(values='temperature', index='date', columns='Country')
df_t_pivot
```
